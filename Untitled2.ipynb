```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
import io
from google.colab import files

# For inline plotting in Colab
%matplotlib inline
```

```python
print("üìÅ Upload your Myntra ZIP file containing the CSV:")
uploaded = files.upload()
```

```python
for filename in uploaded.keys():
    if filename.endswith('.zip'):
        with zipfile.ZipFile(io.BytesIO(uploaded[filename])) as z:
            print("üì¶ Files inside ZIP:", z.namelist())
            csv_file = z.namelist()[0]
            with z.open(csv_file) as f:
                df = pd.read_csv(f, encoding='latin1')
                print("‚úÖ Dataset Loaded Successfully!")
```

```python
print("\nüìä Preview First 5 Rows:")
print(df.head())

print("\nüîé Dataset Info:")
print(df.info())
```

```python
# Correct column names from your dataset
df_cleaned = df.dropna(subset=['name', 'price', 'mrp', 'discount', 'seller'])
```

```python
df.rename(columns={
    'name': 'product_name',
    'mrp': 'original_price',
    'discount': 'discount_percent'
}, inplace=True)
```

```python
df_cleaned = df.dropna(subset=['product_name', 'price', 'original_price', 'discount_percent', 'seller'])
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set visual style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# 1. Top 10 most expensive products
top_expensive = df_cleaned.sort_values(by='original_price', ascending=False).head(10)
sns.barplot(x='original_price', y='product_name', data=top_expensive, palette="coolwarm")
plt.title("Top 10 Most Expensive Products")
plt.xlabel("Original Price")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()

# 2. Discount Distribution
sns.histplot(df_cleaned['discount_percent'].dropna(), bins=20, kde=True, color='orange')
plt.title("Discount Percent Distribution")
plt.xlabel("Discount Percent")
plt.ylabel("Frequency")
plt.show()

# 3. Average Price by Seller (Top 10 sellers by count)
top_sellers = df_cleaned['seller'].value_counts().head(10).index
avg_price_sellers = df_cleaned[df_cleaned['seller'].isin(top_sellers)].groupby('seller')['price'].mean().sort_values(ascending=False)

sns.barplot(x=avg_price_sellers.values, y=avg_price_sellers.index, palette="viridis")
plt.title("Average Price by Top 10 Sellers")
plt.xlabel("Average Price")
plt.ylabel("Seller")
plt.show()

# 4. Rating Distribution
sns.histplot(df_cleaned['rating'].dropna(), bins=20, kde=True, color='green')
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# 5. Correlation Heatmap (Numerical columns only)
numerical_cols = ['price', 'original_price', 'discount_percent', 'rating', 'ratingTotal']
corr = df_cleaned[numerical_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
top_seller_counts = df_cleaned['seller'].value_counts().head(10)

sns.barplot(x=top_seller_counts.values, y=top_seller_counts.index, palette='magma')
plt.title("Top 10 Sellers by Product Count")
plt.xlabel("Number of Products")
plt.ylabel("Seller")
plt.show()
```

```python
price_rating_trend = df_cleaned.groupby('price')['rating'].mean().reset_index().sort_values(by='price')

sns.lineplot(data=price_rating_trend, x='price', y='rating', color='blue')
plt.title("Average Rating across Price")
plt.xlabel("Price")
plt.ylabel("Average Rating")
plt.grid(True)
plt.show()
```

```python
high_rating_discount = df_cleaned[(df_cleaned['rating'] >= 4.5) & (df_cleaned['discount_percent'] >= 50)]
top_picks = high_rating_discount.sort_values(by='discount_percent', ascending=False).head(10)

sns.barplot(x='discount_percent', y='product_name', data=top_picks, palette='crest')
plt.title("Top Rated Products with High Discounts")
plt.xlabel("Discount Percent")
plt.ylabel("Product Name")
plt.show()
```

```python
df_cleaned['price_range'] = pd.cut(df_cleaned['price'], bins=[0, 500, 1000, 2000, 5000, 10000], labels=['<500', '500-1K', '1K-2K', '2K-5K', '5K+'])

sns.boxplot(x='price_range', y='rating', data=df_cleaned, palette='Set2')
plt.title("Rating Distribution by Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Rating")
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(df_cleaned['product_name'].dropna())
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Frequent Words in Product Names", fontsize=16)
plt.show()
```
